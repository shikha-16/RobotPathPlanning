#### Obstacles using polygons

```python
def obst(arr):
    ans = list()
    for coord in arr:
        m = Polygon(coord)
        x,y = m.exterior.xy
        plt.plot(x,y,"black")
        ans.append(m)
    t = MultiPolygon(ans)
    return t
obst([[(1,1),(2,1),(2,2)]])
plt.show()
```

#### Defining Goal region

```python
def goal(coord):
    m = Polygon(coord)
    x,y = m.exterior.xy
    plt.plot(x,y,"black")
    t = MultiPolygon([m])
    return t
    
goal([(6,6),(6,7),(7,7),(7,6)])
plt.show
```
